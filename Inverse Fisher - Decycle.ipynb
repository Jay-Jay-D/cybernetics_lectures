# Testing several Inverse Fisher configurations.

```python
%matplotlib inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 14, 7
```

```python
# First import useful tools
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')

import elhers_indicators as ei
```

```python
data = pd.read_csv('https://drive.google.com/uc?export=download&id=0B9-kA56h5JCMbXptY0kyVXZXaUU',
                 index_col=0, parse_dates=True)
#data = pd.read_csv('https://drive.google.com/uc?export=download&id=0B9-kA56h5JCMU2w4d3JmN1hWZW8',
#                 index_col=0, parse_dates=True)
time = data.index
price = data.CLOSE
```

```python
Decycle = ei.decycle(price, 10)
IFD_st_10 = ei.fisher_inverse(pd.Series(Decycle), 10)
IFD_st_30 = ei.fisher_inverse(pd.Series(Decycle), 30)
IFD_st_45 = ei.fisher_inverse(pd.Series(Decycle), 45)

fig, axs = plt.subplots(2,1,sharex=True)

axs[0].plot(time, price, label='Price')
axs[0].plot(time, Decycle, label='Deycle(10)')
axs[0].legend(loc='upper left')
axs[0].grid()
axs[1].plot(time, IFD_st_10, label='Inverse Fisher Decycle(10) - 10')
axs[1].plot(time, IFD_st_30, label='Inverse Fisher Decycle(10) - 30')
axs[1].plot(time, IFD_st_45, label='Inverse Fisher Decycle(10) - 45')
axs[1].legend(loc='upper right')
axs[1].grid()
```

```python
Decycle = ei.decycle(price, 30)
IFD_st_10 = ei.fisher_inverse(pd.Series(Decycle), 10)
IFD_st_30 = ei.fisher_inverse(pd.Series(Decycle), 30)
IFD_st_45 = ei.fisher_inverse(pd.Series(Decycle), 45)

fig, axs = plt.subplots(2,1,sharex=True)

axs[0].plot(time, price, label='Price')
axs[0].plot(time, Decycle, label='Deycle(30)')
axs[0].legend(loc='upper left')
axs[0].grid()
axs[1].plot(time, IFD_st_10, label='Inverse Fisher Decycle(30) - 10')
axs[1].plot(time, IFD_st_30, label='Inverse Fisher Decycle(30) - 30')
axs[1].plot(time, IFD_st_45, label='Inverse Fisher Decycle(30) - 45')
axs[1].legend(loc='upper right')
axs[1].grid()
```
